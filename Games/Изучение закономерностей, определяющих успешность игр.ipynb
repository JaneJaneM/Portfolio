# Изучение закономерностей, определяющих успешность игр




Интернет-магазин «Стримчик»  продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Перед нами данные до 2016 года. Необходимо спланировать успешную рекламную кампанию на 2017-й, делая ставку на потенциально популярный продукт.


**Цель исследования** - выявить определяющие успешность игры закономерности и на основании этих паремтров спланировать рекламную кампанию на будущий год. 

**Ход исследования:**

Данные для исследования мы получили из файла games.csv. О качестве данных ничего не известно. Таким образом, исследование пройдёт в четыре этапа:

1. Обзор данных.

2. Предобработка данных.

3. Исследовательский анализ данных.

4. Статистический анализ данных.

**Задачи исследования:**

1. Изучить общую информацию о данных, выполнить их предобработку.
2. Проверить данные на пропуски, на ошибки и оценить их влияние на исследование.
3. Провести исследовательский анализ данных.
4. Проанализировать выпуск игр по годам, выделить платформы с наибольшими продажами.
5. Изучить взаимосвязь продаж игр и отзывов критиков и пользоватлей. 
6. Выявить самые прибыльные жанры игр. 
7. Составить портрет пользователя каждого региона, выделив популярные платформы и жанры в разных странах.


**Гипотезы исследования:**
1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
2. Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

**Классификация рейтинга от организации ESRB**

Согласно [википедии](https://ru.wikipedia.org/wiki/Entertainment_Software_Rating_Board) выделяют следующие виды рейтинга игр от ESRB:

«EC» («Early childhood») — «Для детей младшего возраста»: Игра подходит для детей от 3 лет и не содержит материалов, которые родители могли бы счесть неподходящими. Продукты, получившие данный рейтинг, изначально разрабатываются для детей и обычно представляют собой развивающие игры.
  
  «E» («Everyone») — «Для всех»: Содержание вполне подходит для всех возрастов. Такие игры могут понравиться и взрослым. Игры с этим рейтингом могут содержать минимальное насилие, в основном «мультяшного» характера. 
  
  "K-A" ("Kids to Adults"), позже «E10+» («Everyone 10 and older») — «Для всех от 10 лет и старше»: Проекты с данным рейтингом могут содержать больше мультипликационного или мягкого насилия, или несколько откровенные сцены, или минимальное количество крови. Рейтинг был принят ESRB 2 марта 2005 года.
  
  «T» («Teen») — «Подросткам»: Игра подходит для лиц от 13 лет. Проекты из данной категории могут содержать насилие, непристойные сцены, грубый юмор, в меру откровенное сексуальное содержимое, кровь или нечастое использование ненормативной лексики.
  
  «M» («Mature») — «Для взрослых»: Материалы игры не подходят для лиц младше 17 лет. Проекты с данным рейтингом могут содержать жестокое насилие, большое количество крови с расчленением, непристойные сексуальные сцены или грубую ненормативную лексику.
  
  «AO» («Adults Only 18+») — «Только для взрослых»: Содержание игры только для взрослых старше 18 лет. Продукты из данной категории могут содержать длительные сцены жестокого насилия и/или очень откровенное сексуальное содержимое.
  
  «RP» («Rating Pending») — «Рейтинг ожидается»: Продукт был отправлен в ESRB и ожидает присвоения рейтинга. Данный логотип используется только на рекламных презентациях и в демо-версиях игр до официальной даты выпуска в продажу.
  
  «RP 17+» («Rating Pending») — «Рейтинг ожидается - примерно 17+»: Продукт отправлен ESRB на оценку, но примерное содержание близко к рейтингу «M». Данная маркировка была введена с 2021 г. и распространяется на демо или презентационный вариант

Анализируя доли пропусков и абсолютные значения количества пропусков примем решение о удалении, замене или игнорировании пропусков.

В столбце `name` и столбце `genre` - название игры и жанр игры пропущено по 2 значения. Посмотрим на них:

```python
df[df['name'].isna()].head()  # с помощью метода isna() найдем пропуски и выведем на экран строки с проопусками
```

Пропуски в `name` и `genre` совпали. Это 2 игры платформы GEN, 1993 год. Такжен отсутствуют оценки критиков и пользователей. Причины пропусков не ясны, поэтому принято решение удалить эти строки, так как параметр жанр важен для исследования, доля пропусков мала  и в большинстве задач удаление такой доли строк не повлияет на результаты. Также мы не можем чем-то корректным заменить NaN в этом случае. 

```python
df.dropna(subset=['name'], inplace=True)  # удаление строк с пропуском в столбце name
```

В столбце `year_of_release` 1.6% пропусков, удалим их. Заменить чем-то корректным сложно. Пропуски могли возникнуть из-за того, что точная дата игры не определена и информацию не внесли. Либо информация исказилась при выгрузке данных и возникли пропуски. 

```python
df.dropna(subset=['year_of_release'], inplace=True)  # удаление строк с пропуском в столбце name
```

Посмотрим колонку `critic_score`, `user_score`, `rating`  и уникальные значения. 


```python
display(df['critic_score'].unique())  # 
display(df['user_score'].unique())
df['rating'].unique()
```

Посмотрим сколько строк со значением 'tbd' в `user_score`:

```python
df.query('user_score == "tbd"')
```

Доля пропусков в колонке `critic_score` очень большая , отстутствует половина измерений параметра. Возможно не все игры проходят оценку критиками, нет единых требований к выпуску игр и этот параметр не обязателен. Удалить такое количество данных нельзя, заменим пропуски на -999, то есть на дефолтное число, чтоб была возможность отсеивать эти заглушки. 

В столбце `user_score` появились значения `tbd` (To Be Determined) , они означаюсь, что ожидается присвоение рейтинга, поэтому можно интерпретировать эту запись как пропуск. Таких строк 2376.  Заменим `tbd` на nan. А потом также поставим заглушки -999.

В колонке `rating` строковые значения, заменим пропуски на Unknown. Данные в таблице собраны за много лет, возможно пропуски означают, что раньше не было процесса присвоения рейтинга или его невозможно присвоить из-за недостаточности оценок.



```python
df['critic_score'] = df['critic_score'].fillna(-999) # замена пропусков в столбце critic_score на -999
df['user_score'] = df['user_score'].replace('tbd', np.NaN)  # замена значения tbd на пропуски
df['user_score'] = df['user_score'].fillna(-999)  # замена пропусков в столбце user_score на -999
df['rating'] = df['rating'].fillna('Unknow')  # замена пропусков в столбце rating на 'Unknow'
```

```python
display(df['critic_score'].unique())
display(df['user_score'].unique())
df['rating'].unique()
```

```python
df.shape[0]  # shape показывает, сколько в датафрейме строк и колонок. 
```

Для обработки пропусков в колонке critic_score и  user_score мы выбрали дефолтное значение -999, так как пропусков очень много и их заполнение каким-либо усредненным значением может сместить наши оценки.

Построим диаграммы рассеяния для нескольких платформ, также построим матрицу корреляций в виде тепловых карт по корреляциям между продажами и оценками пользователей или критиков. 

```python
# создадим цикл для построения диаграмм рассеяния и расчета корреляции по популярным платформам и по каждому виду отзывов.
list_platform = ['PS4', 'XOne', 'WiiU', 'X360', 'PS3', '3DS']  # список наиболее популярные платформ
for pl in list_platform:
    fig, axs = plt.subplots(1, 3, figsize=(18, 6))
    print(f'Корреляция между продажами игр и отзывами критиков и пользователей для платоформы {pl}')
    sns.scatterplot(data=filter_df.query('(platform == @pl & critic_score >= 0)'),\
                    x='critic_score',\
                    y='total_sales',ax = axs[0])
    sns.scatterplot(data=filter_df.query('(platform == @pl & user_score >= 0)'),\
                    x='user_score',\
                    y='total_sales',\
                    ax = axs[1])
    sns.heatmap(filter_df.query('(platform == @pl & user_score >= 0)\
    | (platform == @pl & critic_score >= 0)')[['critic_score', 'user_score', 'total_sales']].corr(), annot=True, fmt='.3g', ax = axs[2])
    plt.show()
```

Диаграммы рассеяния и матрица корреляций связи продаж и оценками пользователей и критиков показывают, что в целом нет зависимости между отзывами пользователей и продажами. Только у платформы 3DS оценки пользователей связаны с общими продажами, но связь слабая.

### Статистический анализ данных

#### Проверка гипотезы 1:

Н0: Средние пользовательские рейтинги платформ Xbox One и PC равны.

Н1: Средние пользовательские рейтинги платформ Xbox One и PC не равны.



Для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам используем t-тест: scipy.stats.ttest_ind (array1, array2, equal_var). Уровень статитсической значимости возьмем 0.05.

```python
xone_score = filter_df[(filter_df['platform']=='XOne') & (filter_df['user_score'] >= 0) & (filter_df['year_of_release']>=2014)]['user_score']
pc_score = filter_df[(filter_df['platform']=='PC') & (filter_df['user_score'] >= 0) & (filter_df['year_of_release']>=2014)]['user_score']

alpha = 0.05

results = st.ttest_ind(xone_score, pc_score, equal_var=False)

print(f'Выборочное среднее пользовательских рейтингов для платформы "XOne": {xone_score.mean()}')
print(f'Выборочное среднее пользовательских рейтингов для платформы "PC": {pc_score.mean()}')
print('p-значение:', results.pvalue)


if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Таким образом, у нас недостаточно оснований отклонить нулевую гипотезу о равенстве средних пользовательских рейтингов платформ Xbox One и PC. Статистически значимых различий не обнаружено. Средние пользовательские рейтинги платформ Xbox One и PC равны.

#### Проверка гипотезы 2:

Н0: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») равны.

Н1: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») не равны.

Для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам используем t-тест: scipy.stats.ttest_ind (array1, array2, equal_var). Уровень статитсической значимости возьмем 0.05.

```python
action_genre = filter_df[(filter_df['genre']=='Action') & (filter_df['user_score'] >= 0) & (filter_df['year_of_release']>=2014)]['user_score']
sports_genre = filter_df[(filter_df['genre']=='Sports') & (filter_df['user_score'] >= 0) & (filter_df['year_of_release']>=2014)]['user_score']

alpha = 0.05

results = st.ttest_ind(action_genre, sports_genre, equal_var=False)

print(f'Выборочное среднее пользовательских рейтнгов жанра "Action": {action_genre.mean()}')
print(f'Выборочное среднее пользовательских рейтнгов жанра "Sports": {sports_genre.mean()}')
print('p-значение:', results.pvalue)


if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
    
    
```

Таким образом,  мы отвергаем нулевую гипотезу о равенстве средних пользовательских рейтингов жанров Action и Sports и принимаем альтернативную гипотезу о том, что средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») не равны.
